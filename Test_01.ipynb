```python
import pandas as pd
import json
from collections import defaultdict

# Sort to get the most recent row per student last
valid_enrollment_df.sort_values(by=['EMPLID', 'STRM'], inplace=True)

# Step 1: Create blocked_courses dict
fail_or_withdraw_grades = ['F', 'W', 'WA']
blocked_courses_dict = (
    valid_enrollment_df[valid_enrollment_df['CRSE_GRADE_OFF'].isin(fail_or_withdraw_grades)]
    .groupby('EMPLID')['DESCR']
    .apply(lambda x: list(set(x)))
    .to_dict()
)

# Step 2: Build course records
valid_enrollment_df['course_record'] = valid_enrollment_df.apply(lambda row: {
    "STRM": row['STRM'],
    "CLASS_NBR": row['CLASS_NBR'],
    "UNT_TAKEN": row['UNT_TAKEN'],
    "DESCR": row['DESCR'],
    "CRSE_ID": row['CRSE_ID'],
    "CRSE_GRADE_OFF": row['CRSE_GRADE_OFF'],
    "COURSE_TITLE_LONG": row['COURSE_TITLE_LONG'],
    "SUBJECT": row['SUBJECT'],
    "CATALOG_NBR": row['CATALOG_NBR'],
    "ACAD_ORG": row['ACAD_ORG'],
    "GPA_Valid": row['GPA_Valid'],
    "Only_Non_GPA_Grades": row['Only_Non_GPA_Grades'],
    "schedule_match": row['schedule_match']
}, axis=1)

# Step 3: Group by student info
grouped = valid_enrollment_df.groupby(['EMPLID', 'NAME_DISPLAY', 'ACAD_CAREER', 'INSTITUTION'])

# Step 4: Assemble final JSON
enrollment_json = defaultdict(list)

for (emplid, name, career, institution), group in grouped:
    latest_row = group.iloc[-1]
    courses = group['course_record'].tolist()
    blocked = blocked_courses_dict.get(emplid, [])

    record = {
        "EMPLID": str(emplid),
        "NAME_DISPLAY": name,
        "ACAD_CAREER": career,
        "INSTITUTION": institution,
        "ACAD_PROG": latest_row['ACAD_PROG'],
        "CUM_GPA": latest_row['CUM_GPA'],
        "blocked_courses": blocked,
        "courses": courses
    }

    enrollment_json[str(emplid)].append(record)

# Step 5: Save as JSON
with open("enrollment_transformed.json", "w", encoding="utf-8") as f:
    json.dump(enrollment_json, f, indent=2)

print("âœ… Saved as 'enrollment_transformed.json'")
```

```python
# Convert columns to numeric
class_schedule_df['Cap Enrl'] = pd.to_numeric(class_schedule_df['Cap Enrl'], errors='coerce')
class_schedule_df['Tot Enrl'] = pd.to_numeric(class_schedule_df['Tot Enrl'], errors='coerce')

# Flags
class_schedule_df['Over_Capacity_Flag'] = (class_schedule_df['Tot Enrl'] > class_schedule_df['Cap Enrl']) & (class_schedule_df['Cap Enrl'] > 0)
class_schedule_df['Zero_Capacity_Enrolled_Flag'] = (class_schedule_df['Cap Enrl'] == 0) & (class_schedule_df['Tot Enrl'] > 0)
class_schedule_df['Full_Capacity_Flag'] = (class_schedule_df['Tot Enrl'] == class_schedule_df['Cap Enrl']) & (class_schedule_df['Cap Enrl'] > 0)

# Capacity issue description
def flag_reason(row):
    if row['Over_Capacity_Flag']:
        return 'Over Capacity'
    elif row['Zero_Capacity_Enrolled_Flag']:
        return 'Zero Capacity with Enrollment'
    elif row['Full_Capacity_Flag']:
        return 'Full'
    else:
        return 'OK'

class_schedule_df['Capacity_Issue'] = class_schedule_df.apply(flag_reason, axis=1)
```

```python
# Flag classes with missing both start and end time
class_schedule_df['Missing_Timing_Flag'] = class_schedule_df['Mtg Start'].isna() & class_schedule_df['Mtg End'].isna()

# Optional: Add explanation column
class_schedule_df['Timing_Issue'] = class_schedule_df['Missing_Timing_Flag'].apply(
    lambda x: 'Missing Timings' if x else 'OK'
)
```
