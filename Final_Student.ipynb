```python
import pandas as pd

# Load your data
# enrollment_df = pd.read_excel("2. 20_STUDENTS_ENROLLEMENT.xlsx")

# Filter electives
elective_orgs = ['UC-CAS', 'UC-COB', 'UC-COE']
electives_df = enrollment_df[enrollment_df['ACAD_ORG'].isin(elective_orgs)]

# Group electives by student
student_electives = electives_df.groupby('EMPLID')['DESCR'].apply(list).reset_index()
student_electives.rename(columns={'DESCR': 'Elective_Courses'}, inplace=True)

# Merge back with original data
enrollment_with_electives = pd.merge(enrollment_df, student_electives, on='EMPLID', how='left')

# Optional: Convert list to string if you need to print/display
enrollment_with_electives['Elective_Courses_finished'] = enrollment_with_electives['Elective_Courses'].astype(str)

# Display the result
# print(enrollment_with_electives[['EMPLID', 'NAME_DISPLAY', 'Elective_Courses_Str']].drop_duplicates().head())
```
